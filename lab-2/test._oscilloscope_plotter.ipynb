```python
import plotly.express as px
import pandas as pd
import numpy as np


file_name = r'NewFile2.csv'

df = pd.read_csv(file_name, header=None)

display(df)



#drop the first 2 rows
df = df.drop([0,1])
display(df)

#drop every column except 0 and 1

df = df.drop([2,3,4], axis=1)
display(df)


#conver CH1 and CH2 to float
df[0] = df[0].astype(float)
df[1] = df[1].astype(float)
display(df)


#put labels on the columns
df.columns = ['CH1', 'CH2']
display(df)
```

```python
increment_seconds = 2.00000e-06
df['Time'] = np.arange(0, increment_seconds*len(df), increment_seconds)
df['Microseconds'] = df['Time']*1e6
```

```python
display(df)
```

```python
scl = 'CH1'
sda = 'CH2'
df[scl] = df[scl] - 4
df[sda] = df[sda] + 4
```

```python
step = 10
step_size = int(len(df)/step)
print(step_size)
fig = px.line(df, x='Microseconds', y= [sda, scl], title='Oscilloscope Plotter')
fig.show()
#add offset fo scl
```

```python
# fig = px.line(df, x='Microseconds', y='CH1')
# fig.show()
# fig = px.line(df, x='Microseconds', y='CH2')
# fig.show()


#plot on 10 steps
step = 10
step_size = int(len(df)/step)
print(step_size)

scl = 'CH1'
sda = 'CH2'


for i in range(step):
    fig = px.line(df.iloc[i*step_size:(i+1)*step_size], x='Microseconds', y='CH1')
    fig.update_layout(
        title="SCL",    
        xaxis_title="Time (microseconds)",
        yaxis_title="Voltage (V)"
    )
    fig.show()
    fig = px.line(df.iloc[i*step_size:(i+1)*step_size], x='Microseconds', y='CH2')
    fig.update_layout(
        title='SDA',
        xaxis_title="Time (microseconds)",
        yaxis_title="Voltage (V)"
    )
    fig.show()
    
fig = px.line(df, x='Microseconds', y= scl)
fig.update_layout(
    title="SCL",    
    xaxis_title="Time (microseconds)",
    yaxis_title="Voltage (V)"
)
fig.show()
fig = px.line(df, x='Microseconds', y= sda)
fig.update_layout(
    title='SDA',
    xaxis_title="Time (microseconds)",
    yaxis_title="Voltage (V)"
)
fig.show()
    
```
