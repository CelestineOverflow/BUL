```python
import pandas as pd
import serial
import plotly.express as px
```

```python
port = 'COM9'
baud = 9600
ser = serial.Serial(port, baud, timeout=1)
```

```python

def read(number_of_lines):
    if not ser.is_open:
        ser.open()
    df = pd.DataFrame(columns=['acc_x', 'acc_y', 'acc_z', 'gyro_x', 'gyro_y', 'gyro_z', 'temp'])
    for i in range(number_of_lines):
        line = ser.readline().decode('utf-8').rstrip()
        coma_count = line.count(',')
        if not coma_count == 6:
            continue
        parts = line.split(',')
        acc_x = int(parts[0])
        acc_y = int(parts[1])
        acc_z = int(parts[2])
        gyro_x = int(parts[3])
        gyro_y = int(parts[4])
        gyro_z = int(parts[5])
        temp = int(parts[6])
        df = pd.concat([df, pd.DataFrame([[acc_x, acc_y, acc_z, gyro_x, gyro_y, gyro_z, temp]], columns=['acc_x', 'acc_y', 'acc_z', 'gyro_x', 'gyro_y', 'gyro_z', 'temp'])], ignore_index=True)
    ser.close()
    return df
```

```python
df = read(1000)
file_name = input('Enter file name: ')
df.to_csv(file_name + '.csv')
fig = px.line(df, x=df.index, y=['acc_x'])
fig.show()
fig = px.histogram(df, x='acc_x')
fig.show()
```
