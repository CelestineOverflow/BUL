```python
import pandas as pd
import numpy as np	
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
files = ['250000.csv', '500000.csv']

# df = pd.read_csv(file, skiprows=20)

df = pd.DataFrame()

for file in files:
    df_temp = pd.read_csv(file, skiprows=20)
    df_temp['file'] = file
    df = df.append(df_temp)
```

```python
display(df)
```

```python
#convert TIME, CH1, CH2 to numeric values
df['TIME'] = pd.to_numeric(df['TIME'], errors='coerce')
df['CH1'] = pd.to_numeric(df['CH1'], errors='coerce')
df['CH2'] = pd.to_numeric(df['CH2'], errors='coerce')

# fig = px.line(df, x='TIME', y=['CH1', 'CH2'])
# fig.show()
#show plot as diferent lines
fig = px.line(df, x='TIME', y=['CH1', 'CH2'], color='file')
fig.show()
```
